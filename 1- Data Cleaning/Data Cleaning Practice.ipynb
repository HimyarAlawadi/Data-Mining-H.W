# Data Cleaning 
I cleaned this dataset in this way based on senario of prediction of real estate price. 

```python
import pandas as pd
```

```python
df = pd.read_csv("./NY-House-Cleaned-Dataset.csv")
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.info()
```

```python
# 1. change BEDS, BATH, from float to int. (inconstant values)
# 2. Remove MAIN_ADDRESS Cloumn. 
# 3. No Massing Value. (no null values)
# 4. Make PROPERTYSQFT as int.
df.drop(labels=None,columns="BROKERTITLE",inplace=True)
df.head()
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
columnsName =["MAIN_ADDRESS","LONGITUDE","LATITUDE","FORMATTED_ADDRESS"]
df.drop(labels=None,columns=columnsName,inplace=True)
df.head()
```

```python
df.drop(labels=None,columns="ADDRESS",inplace=True)
```

```python
df['BATH'] = df['BATH'].astype(int)
df.head()
```

```python
df['PROPERTYSQFT'] = df['PROPERTYSQFT'].astype(int)
df.head()
```

```python
df.drop(labels=None,columns="STATE",inplace=True)
df.head()
```
