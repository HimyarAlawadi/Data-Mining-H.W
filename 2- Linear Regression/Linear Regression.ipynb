# Cleaning Dataset
I cleaned this dataset in this way based on senario of pridect of real estate price in relation to its square footage. 

```python
# import pandas lib to cleaning our dataset
import pandas as pd
```

```python
#Load Dataset
df = pd.read_csv("./NY-House-Cleaned-Dataset.csv")
df.head()
```

```python
# Drop every unwanted column, because they unrelated to our mining scenario
columnsName =["BROKERTITLE","TYPE","BEDS","BATH","ADDRESS","STATE","MAIN_ADDRESS","ADMINISTRATIVE_AREA_LEVEL_2","LOCALITY","SUBLOCALITY","STREET_NAME","LONG_NAME","LONGITUDE","LATITUDE","FORMATTED_ADDRESS"]
df.drop(labels=None,columns=columnsName,inplace=True)
df.head()
```

# Visualize Dataset

```python
# import matplotlib to visualize dataset
import matplotlib.pyplot as plt
```

```python
x = pd.DataFrame(df, columns = ["PROPERTYSQFT"])
y = pd.DataFrame(df, columns = ['PRICE'])

df.describe()
```

```python
plt.scatter(x, y, alpha = 0.3)

plt.xlabel("Property square footage")
plt.ylabel("Price")

plt.xlim(0, 2184.207862)
plt.ylim(0, 1.495000e+06)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x, y, alpha = 0.3)

plt.xlabel("Property square footage")
plt.ylabel("Price")

plt.plot(x, reg.predict(x), color='yellow')

plt.xlim(0,  2184.207862)
plt.ylim(0, 1.495000e+06)
```

As we see above increase proprty square footage price will increase too.
